```python
lines = '''pageInfo:bksMain
login_chk:null
LOGIN_SN:null
LOGIN_NAME:null
indexName:news
keyword:((부산) OR (부산시) OR (부산광역시))
byLine:
searchScope:1
searchFtr:1
startDate:2020-03-26
endDate:2020-06-26
sortMethod:date
contentLength:100
providerCode:
categoryCode:
incidentCode:
dateCode:
highlighting:true
sessionUSID:
sessionUUID:test
listMode:
categoryTab:
newsId:
filterProviderCode:
filterCategoryCode:
filterIncidentCode:
filterDateCode:
filterAnalysisCode:
startNo:1
resultNumber:39675
topmenuoff:
resultState:
keywordJson:
keywordFilterJson:
realKeyword:
totalCount:
interval:
quotationKeyword1:
quotationKeyword2:
quotationKeyword3:
searchFromUseYN:N
mainTodayPersonYn:
period:3month'''.splitlines()

data = {}
for line in lines:
    key, value = line.split(':', 1)
    if value == 'null':
        value = None
    data[key] = value

import requests
from bs4 import BeautifulSoup
import pandas as pd

result_url = "https://www.bigkinds.or.kr/news/newsResult.do"
response = requests.post(result_url, data=data)
html = response.text
soup = BeautifulSoup(html, 'html.parser')

title=[]
url=[]

for tag in soup.select('.resultList li h3'):
    doc_id = tag['id'].replace('news_', '')
    doc_url = 'https://www.bigkinds.or.kr/news/detailView.do?docId={}&returnCnt=1'.format(doc_id)
    title.append(tag.text.strip())
    url.append(doc_url)
    
import json
import urllib.request
date=[]
writer=[]
category=[]
content=[]
for i in range(len(url)):
    html=urllib.request.urlopen(url[i]).read()
    soup=BeautifulSoup(html, 'html.parser')
    site_json=json.loads(soup.text)
    detail=site_json['detail']
    value=detail["DATE"]
    date.append(value)
    w=detail['BYLINE']
    writer.append(w)
    c=detail['CATEGORY_MAIN']
    category.append(c)
    con=detail['CONTENT']
    content.append(con)

DF=pd.DataFrame({'title':title, 'date':date, 'writer':writer, 'category':category, 'content':content,'json url':url})
DF
```

```python
DF.to_csv('/Users/ourlocalpetition/Desktop/bigkinds.csv', header=True, index=False,encoding='utf-8-sig')
```
